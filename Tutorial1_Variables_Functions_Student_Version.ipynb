In order to access the output of a function, you must call the function by using its name followed by parentheses:

````python
#defining variables
ctrl_mice = [60, 126, 50, 300, 421]
insomniac_mice = [10, 1, 20, 30, 42]

def sleep_amounts_calc_percent_diff(ctrl_mice, insomniac_mice): #definining a function
  avg_control_mice = np.mean(ctrl_mice) #calculating average for control and experimental mice
  avg_exp_mice = np.mean(insomniac_mice)
  percent_diff_between_groups = ((avg_exp_mice - avg_control_mice)/avg_control_mice)*100 #calculating percent difference between two groups
  return percent_diff_between_groups #returning an output

sleep_amounts_calc_percent_diff(ctrl_mice, insomniac_mice) #calling this function
print(sleep_amounts_calc_percent_diff(ctrl_mice, insomniac_mice))
````

# **IF Statements:**

You can see that simply going through the list with a loop is not that helpful on its own. Yes, it goes through each state one by one—but what if you only want to isolate and work with **NREM (NR) sleep** values? This is where conditionals come in. By combining a loop with an if statement, we can tell Python to only do something when a specific condition is met—like when the state is **"NR"**.

1131424.png

````python
Mouse_Sleep_Wake_Amounts = ["NR", "NR", "NR", "R", "R", "R", "W", "W", "W", "NR", "NR", "NR", "NR", "NR", "W", "NR"]

#creating a for-loop
for sleep_wake_state in Mouse_Sleep_Wake_Amounts: #sleep_wake_state is temporally element in the list.
  if .. == "NR": #check if the temporally element meet a condition
    print(sleep_wake_state)
````

# **⚠️ Notice:**

We used **==** inside the conditional statement. If you try to use a single **=** instead, Python will give you an error.

That’s because:

A **single =** is used to assign a value to a variable.

A **double ==** is used to check whether two values are equal.

So when you're writing an if statement to compare something—like checking if the current state is "NR"—you need to use "==", **not** "=".

````python
Mouse_Sleep_Wake_Amounts = ["NR", "NR", "NR", "R", "R", "R", "W", "W", "W", "NR", "NR", "NR", "NR", "NR", "W", "NR"]

#creating a for-loop
for sleep_wake_state in Mouse_Sleep_Wake_Amounts: #sleep_wake_state is temporally element in the list.
  if sleep_wake_state = "NR": #check if the temporally element meet a condition
    print(sleep_wake_state)
````

# **Multiple Conditions with elif and else**

To summarize:



*   **if** checks the first condition

Sometimes, you would want to iterate through data and count it only **if it does not equal a certain value.** You can simply use **!=** to do that.
